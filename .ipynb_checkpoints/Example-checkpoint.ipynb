```python
import requests, re
from bs4 import BeautifulSoup as bs
import pandas as pd, numpy as np
from tqdm import tqdm
```

# MonexNew

```python
url = 'https://monexnews.com/kurs-valuta-asing.htm?kurs=JPY&searchdatefrom=12-05-2019&searchdateto=12-05-2020'
res = requests.get(url)
```

```python
soup = bs(res.content, 'html.parser')
table = soup.find('table', {'class':'table'})
rows = table.find_all('tr')
_data = []
for row in rows:
    _data.append([col.text for col in row.find_all('td')])

df = pd.DataFrame(_data[1::], columns=['Date', 'Ask', 'Bid'])
```

```python
df['Ask'] = df['Ask'].str.replace(',', '.')
df['Bid'] = df['Bid'].str.replace(',', '.')
df[['Ask', 'Bid']] = df[['Ask', 'Bid']].astype(float)
```

Dari website imbd, setiap page review memiliki total 50 movie. Maka, setiap page akan berulang untuk kelipatan 50. Kita akan coba otomasikan untuk pencarian movie lebih dari 50 (lebih dari 1 page)

```python
def scrap(url, movie_count, threshold):
    res = requests.get(url) #get url
    soup = bs(res.content) 
    
    # Buat container data
    data = {'title':[], 'review_score':[], 'meta_score':[], 'votes':[],
            'director':[], 'artist': [],'genre':[], 'duration':[], 'synopsis': []
           }
        
    # Ambil semua(50) review dalam 1 page
    reviews = soup.find_all('div', {'class':'lister-item mode-advanced'})
    
    for movie in reviews:
        
        # set variabel yang mungkin kosong. males bikin try-catch satu-satu. (jangan ditiru)
        
        
        title = movie.find('h3', {'class': 'lister-item-header'}).find('a').text
        try : 
            review = float(movie.find('div', {'class':'inline-block ratings-imdb-rating'}).find('strong').text)
        except : 
            review = np.nan
        try : 
            votes = int(movie.find('span', {'name':'nv'})['data-value'])
        except : 
            votes = 0
        try :
            meta_score = int(movie.find('span', {'class': 'metascore mixed'}).text)
        except:
            meta_score = np.nan
        try: 
            info = movie.find_all('p', {'class': 'text-muted'})
            person = movie.find('p', {'class': ''})
            
            # get duration, synopsis and genre
            duration = int(re.findall(r'(\d+)', info[0].find('span', {'class': 'runtime'}).text)[0])
            genre = info[0].find('span', {'class':'genre'}).text.replace('\n', '')
            synopsis = info[1].text

            # get artists and director
            separator = person.text.find('|') 
            if separator != -1: # ada director
                idx_start = person.text.find('Director:')
                director = person.text[idx+10:separator].replace('\n', '')#.split(',')
            else:
                director = np.nan
            artist = person.text[person.text.find('Stars')+7:].replace('\n', '')
        except:
            # Notes: ini lagi males nulis try-catch utuk setiap variabel. Jangan ditiru. 
            duration = np.nan
            genre = np.nan
            synopsis = np.nan
            artist = np.nan 
            director = np.nan
        
        
        data['title'].append(title)
        data['review_score'].append(review)
        data['meta_score'].append(meta_score)
        data['votes'].append(votes)
        data['director'].append(director)
        data['artist'].append(artist)
        data['genre'].append(genre)
        data['duration'].append(duration)
        data['synopsis'].append(synopsis)

        movie_count += 1
        if movie_count == threshold :
            break

    df = pd.DataFrame(data)
    return df, movie_count
```
